## Problem 2.1
Recall that the solution is $u(x) = \frac{\exp\left(\frac{Pe\,x}{L}\right) - 1}{\exp(Pe) - 1}$
\begin{align}
|| u||^2_{H^2} &= \int_0^L \big(u^2 + L^2 u_{,x}^2 + L^4 u_{,xx}^2 \big)\, dx\\
&= \frac{1}{(\exp(Pe) - 1)^2} \int_0^L \left(\left(\exp\left(\frac{Pe\,x}{L}\right)-1 \right)^2 + Pe^2\exp\left(\frac{2Pe\,x}{L}\right) + Pe^4\exp\left(\frac{2Pe\,x}{L}\right)\right)\, dx\\
&\sim \frac{1}{\exp(2Pe)}\int_0^L Pe^4 \exp\left(\frac{2Pe\,x}{L}\right)\, dx \,\quad \text{ for large } Pe\\
&= \frac{Pe^4}{\exp(2Pe)}\left(\frac{L}{Pe} \big(\exp(2Pe)-1 \big) \right) \,\sim\, Pe^3 L
\end{align}



## Problem 2.3

Because $u\in C^2([0,L])$, for $x\in[x_A, x_{A+1}]$ we have $$u(x) = u(x_A)+u'(x_A)(x-x_A) + \frac{u''(\xi(x))}{2}(x-x_A)^2\quad\_\_\_\,(1)$$ for some $\xi(x) \in [x_A, x_{A+1}]$ between $x$ and $x_A$. 

Also note that the linear interpolate of $u,\,$ $\breve{u}^h(x) = u(x_A) +\frac{u(x_{A+1}) - u(x_A)}{x_{A+1} - x_A}(x-x_A) \,$ for $x\in[x_A, x_{A+1}]$, is such that , $\breve{u}^h(x_A)=u(x_A)\,$ and $\,{\breve{u}^h}_{,x}(x) = \frac{u(x_{A+1}) - u(x_A)}{x_{A+1} - x_A}$ 

Applying (1) to
this last expression gives $$\breve{u}^h_{,x} = u'(x_A) + \frac{u''(\xi(x_{A+1}))}{2}\underbrace{(x_{A+1}-x_A)}_{:= \,h_A}$$  


Thus,
\begin{align}
||\eta||_{L^\infty} &= \inf_{v^h\in V^h} ||u - v^h||_{L^{\infty}}\le  ||u - \breve{u}^h||_{L^\infty} \\ &= | u(y) - \breve{u}^h(y)|\quad \text{ for some } y\in[0,L]\quad (y\in [x_A, x_{A+1}] \text{ for some } A)\\
&\le \left|u'(x_A) - \frac{u(x_{A+1}) - u(x_A) }{h_A}\right| \,|x-x_A| + \frac{u''(\xi(y))}{2}(x-x_A)^2\\
&= \left|\frac{u''(\xi(x_{A+1})}{2}\right| h_A|x-x_A| + \frac{|u''(\xi(y))|}{2}(x-x_A)^2 \\
&\le \frac{1}{2}\big(h_A|x-x_A| + (x-x_A)^2)\big)\, ||u'' ||_{L^\infty} \\ \\
&\le \frac{1}{2} h_A^2 \,||u'' ||_{L^\infty} 
\end{align}

The estimate $||\eta_{,x} ||_{L^\infty}\le \frac{1}{2} h_A || u''||_{L^\infty}$ can be obtained using $| u_{,x}(y) - \breve{u}^h_{,x}(y)|$ in place of $| u(y) - \breve{u}^h(y)|$ and following the same steps as above.
A more careful analysis can produce a bounding constant smaller than $\frac{1}{2}$.

## Problem 2.4

Without loss of generality, assume $\Omega^e = [0,\,h]$ and notice that 
\begin{align}
\frac{|| w^h_{,x}||^2_{\Omega^e}}{||w^h ||^2_{\Omega^e}} &= \frac{\sum_A \sum_B\, w_A\Big( \int_0^h N_{A,x} N_{B,x}\,dx \Big) \, w_B }{\sum_A \sum_B\, w_A\Big( \int_0^h N_{A} N_{B}\,dx \Big) \, w_B} = \frac{1}{h^2}\frac{\boldsymbol{w}^T\boldsymbol{K}\boldsymbol{w}}{\boldsymbol{w}^T\boldsymbol{M}\boldsymbol{w}}
\end{align}

where  $\,\boldsymbol{K}_{AB} = h\int_0^h N_{A,x} N_{B,x}dx\,$ and $\,\boldsymbol{M}_{AB} = \frac{1}{h}\int_0^h N_{A} N_{B}dx\, $. Notice that the above ratio takes the form of a generalized Rayleigh quotient and is thus bounded by the maximum and minimum eigenvalues in $\,(\boldsymbol{K} - \lambda\boldsymbol{M})\,\boldsymbol{x} = 0$. 

Thus calculating $C_{inv}$ amounts to finding the maximum eigenvalue for this problem.

#### Piecewise linears
We have two bases: $N_1(x) = \frac{x}{h}$ and $N_2(x) = 1 - \frac{x}{h}$  which give
\begin{align}
&\boldsymbol{K} = \begin{pmatrix}1 & -1\\ -1 & 1  \end{pmatrix}  & \boldsymbol{M} = \frac{1}{6}\begin{pmatrix}2 & 1\\ 1 & 2  \end{pmatrix}
\end{align}
from which one can show that $\lambda_{\text{max}} = 12$. Thus $$ \frac{|| w^h_{,x}||^2_{\Omega^e}}{||w^h ||^2_{\Omega^e}} \le \frac{12}{h^2}\quad \Leftrightarrow\quad || w^h_{,x}||_{\Omega^e} \le \frac{\sqrt{12}}{h}\,|| w^h||_{\Omega^e}$$

#### Piecewise quadratics
$N_1(x) = 2\big(\frac{x}{h} - \frac{1}{2}\big) \big(\frac{x}{h}-1 \big)$, $\,N_2(x) = \frac{4x}{h} \big(1-\frac{x}{h} \big)\,$ and $\,N_3(x) = \frac{2x}{h}  \big(\frac{x}{h}-\frac{1}{2} \big)$

\begin{align}
&\boldsymbol{K} = \frac{1}{3}\begin{pmatrix}7 & -8 & 1 \\ -8 & 16 & -8 \\ 1& -8& 7 \end{pmatrix}  & \boldsymbol{M} = \frac{1}{30}\begin{pmatrix}4 & 2 & -1\\ 2 & 16 & 2 \\ -1 & 2 & 4  \end{pmatrix}
\end{align}
which gives $\lambda_{\text{max}} = 60$ or $C_{inv} =\sqrt{60}$

#### Piecewise quadratics (part 2)
Notice that $|| w^h_{,xx}||_{\Omega^e} \le \frac{C_{inv}}{h}\,|| w^h_{,x}||_{\Omega^e}$ for quadratics is equivalent to $|| w^h_{,x}||_{\Omega^e} \le \frac{C_{inv}}{h}\,|| w^h||_{\Omega^e}$ for linears. Thus $C_{inv} = \sqrt{12}$

#### Piecewise quadratics (part 3)
 $|| w^h_{,xx}||_{\Omega^e} \le \frac{\sqrt{12}}{h}\,|| w^h_{,x}||_{\Omega^e} \le \frac{\sqrt{12}}{h} \frac{\sqrt{60}}{h}||w^h|| = \frac{\sqrt{720}}{h^2}||w^h||$

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
import scipy.sparse as spsp
import scipy.sparse.linalg as spspla
%matplotlib notebook
```

```python
def XiFunction(h, a, kappa): 
    if kappa == 0:
        return 1
    elif a == 0:
        return 0
    else:
        alpha = a * h/ (2 * kappa)
        return 1/np.tanh(alpha) - 1/alpha
    
def KmatrixAssemble(N=11, a=1, kappa=1, XiFlag='SUPG'):
    h = 1/ N
    #alpha = alphaFunction(h, a, kappa)
    
    if XiFlag == 'SUPG':
        Xi = XiFunction(h, a, kappa)
    elif XiFlag ==  'Galerkin':
        Xi = 0
    else:
        Xi=0
        
    DiffusionStencil = np.array([-1, 2, -1 ])
    #print(DiffusionStencil.shape)
    AdvectionStencil = np.array([-1, 0, 1])
    
    K = np.zeros((N-1, N-1))
    K[0, 0:2] = (kappa / h + a * Xi / 2) *np.array([2, -1]) +  (a / 2) *np.array([0, 1])
    K[-1, -2:] = (kappa / h + a * Xi / 2) *np.array([-1, 2]) +  (a / 2) *np.array([-1, 0])
    for A in range(1,N-2):
        K[A, A-1:A+2] = (kappa / h + a * Xi / 2) * DiffusionStencil + (a / 2) * AdvectionStencil
    
    return spsp.csr_matrix(K)

def leftBoundaryVector(N=11, a=1, kappa=1, XiFlag='SUPG'):
    h = 1/ N
    #alpha = alphaFunction(h, a, kappa)
    
    if XiFlag == 'SUPG':
        Xi = XiFunction(h, a, kappa)
    elif XiFlag ==  'Galerkin':
        Xi = 0
    else:
        Xi=0
    
    B0 = np.zeros(N-1)
    B0[0] = (kappa / h + a * Xi / 2) *np.array([-1]) +  (a / 2) *np.array([-1])
    
    return B0

def rightBoundaryVector(N=11, a=1, kappa=1, XiFlag='SUPG'):
    h = 1/ N
    #alpha = alphaFunction(h, a, kappa)
    
    if XiFlag == 'SUPG':
        Xi = XiFunction(h, a, kappa)
    elif XiFlag ==  'Galerkin':
        Xi = 0
    else:
        Xi=0
    
    BN = np.zeros(N-1)
    BN[-1] = (kappa / h + a * Xi / 2) *np.array([-1]) +  (a / 2) *np.array([1])
    
    return BN  


def USolve(g0, g1, N=11, a=500, kappa=1, XiFlag='SUPG'):
    K = KmatrixAssemble(N=N, a=a, kappa=kappa, XiFlag = XiFlag)
    B0 = leftBoundaryVector(N=N, a=a, kappa=kappa, XiFlag=XiFlag)
    BN = rightBoundaryVector(N=N, a=a, kappa=kappa, XiFlag=XiFlag)
    
    U = np.zeros(N+1)
    U[0]= g0
    U[-1] = g1
    U[1:-1] = spspla.spsolve(K, -g0*B0 - g1*BN )
    
    return U

def Basis(A, x, N=11):
    xA = A/N
    xA1 = (A+1)/N
    xAm1 = (A-1)/N
    parentN = lambda x: 1 - np.abs(x)
    
    if x <= xA1 and x > xAm1:
        return parentN( N*x - A )
    else:
        return 0

def DBasis(A, x, N=11):
    xA = A/N
    xA1 = (A+1)/N
    xAm1 = (A-1)/N
    
    if x <= xA and x > xAm1:
        return N
    elif x > xA and x < xA1:
        return -N
    else:
        return 0
    
def Ufem(U, x):
    U_of_x = 0
    for i in range(len(U)):
        U_of_x += U[i] * Basis(i, x, N=len(U)-1)
    return U_of_x

def dUfem(U, x):
    U_of_x = 0
    for i in range(len(U)):
        U_of_x += U[i] * DBasis(i, x, N=len(U)-1)
    return U_of_x

def Uexact(x, a=500, kappa=1):
    if a == 0:
            Ux = x
    else:
            Ux = (np.exp(a * x / kappa) - 1) / (np.exp(a / kappa) - 1)  
    return Ux

def dUexact(x, a=500, kappa=1):
    if a == 0:
            dUx = 1
    else:
            dUx = (a/kappa) * np.exp(a * x / kappa) / (np.exp(a / kappa) - 1)  
    return dUx

def L2error(U):
    L2sqr = integrate.quad(lambda x: (Uexact(x) - Ufem(U, x))**2, 0, 1, limit=75,\
                           points=[.95,.97,.98,.98125,.9825,.985,.9875,.98875, .99,.9925,.995,.999, .9995])[0]
    return np.sqrt(L2sqr)

def H1semierror(U):
    H1semisqr =integrate.quad(lambda x: (dUexact(x) - dUfem(U, x))**2, 0, 1, limit=75,\
                             points=[.95,.97,.98,.98125,.9825,.985,.9875,.98875,.99,.9925,.995,.999, .9995])[0]
    return np.sqrt(H1semisqr)

def H1error(U):
    return np.sqrt(L2error(U)**2 + H1semierror(U)**2)
```

```python
Ns = np.array([3, 5, 10, 25, 50, 100, 250, 500, 1000])

#Galerkin
L2Galerkin = np.zeros(len(Ns))
H1semiGalerkin = np.zeros(len(Ns))
H1Galerkin = np.zeros(len(Ns))

for i in range(len(Ns)):
    UGal = USolve(0,1, N = Ns[i], XiFlag= 'Galerkin')
    L2Galerkin[i]= L2error(UGal)
    H1semiGalerkin[i] = H1semierror(UGal)
    H1Galerkin[i]= H1error(UGal)

    
#SUPG
L2SUPG = np.zeros(len(Ns))
H1semiSUPG = np.zeros(len(Ns))
H1SUPG = np.zeros(len(Ns))

for i in range(len(Ns)):
    USUPG = USolve(0,1, N = Ns[i], XiFlag= 'SUPG')
    L2SUPG[i]= L2error(USUPG)
    H1semiSUPG[i] = H1semierror(USUPG)
    H1SUPG[i]= H1error(USUPG)
#print(L2SUPG)
```
